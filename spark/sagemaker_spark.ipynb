```python
import pandas as pd

# Sample DataFrame
data = {
    'Column1': ['0', '1', '2', '3'],
    'Column2': ['4', '5', '0', '6'],
    'Column3': ['7', '0', '8', '9']
}

df = pd.DataFrame(data)

# Function to identify string 'zero'
def find_string_zeros(value):
    return value == '0'

# Loop through each column and apply the function
columns_with_string_zeros = []
for column in df.columns:
    # Use .applymap() to apply the function to each cell in the column
    if df[column].applymap(find_string_zeros).any():
        columns_with_string_zeros.append(column)

# Display columns with string 'zero'
print("Columns with string 'zero':", columns_with_string_zeros)
```
